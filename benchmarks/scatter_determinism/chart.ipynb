```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("results.csv")
plt.plot(data["tensor_size"], data["cpu"] * 100, marker="o", linestyle="--")
plt.plot(data["tensor_size"], data["cuda"] * 100, marker="o", linestyle="--")
plt.legend(["CPU", "CUDA"])
plt.xscale("log")
plt.xlabel("Rozmiar tensora")
plt.ylabel("Procent prb, w ktrych zastpowanie\nzostao wykonane po kolei")
plt.savefig("bench_scatter_determinism.pdf")
```
