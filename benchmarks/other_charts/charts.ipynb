```python
import torch
import matplotlib.pyplot as plt


year = [2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
pytorch = torch.tensor(
    [
        801,
        4820,
        14300,
        31700,
        49600,
        48400,
        41600,
        24400,
    ]
)
tensorflow = torch.tensor(
    [
        8230,
        22200,
        36800,
        47600,
        49700,
        42100,
        33500,
        14800,
    ]
)


percentage = (pytorch / (pytorch + tensorflow)) * 100

plt.plot(year, percentage, color="black", label="_nolegend_")
plt.fill_between(year, [0] * len(year), percentage)
plt.fill_between(year, percentage, [100] * len(year))
plt.xlim(year[0], year[-1])
plt.ylim(0, 100)
plt.yticks([0, 25, 50, 75, 100])
yticks = [0, 25, 50, 75, 100]
plt.yticks(yticks, ["{:.0f}%".format(x) for x in yticks])
plt.legend(("PyTroch", "TensorFlow"))
plt.grid()
fig = plt.gcf()
fig.set_size_inches(8, 4)
plt.savefig("pytorch_vs_tensorflow.pdf")
```
