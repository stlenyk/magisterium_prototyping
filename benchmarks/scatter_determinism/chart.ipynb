```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("results.csv")
# add lines to datapoints
plt.plot(data["tensor_size"], data["cpu"], marker="o", linestyle="--")
plt.plot(data["tensor_size"], data["cuda"], marker="o", linestyle="--")
plt.legend(["CPU", "CUDA"])
plt.xscale("log")
plt.xlabel("Rozmiar tensora")
plt.ylabel("Uamek prb, w ktrych zastpowanie\nzostao wykonane po kolei")
plt.savefig("bench_scatter_determinism.pdf")
```
